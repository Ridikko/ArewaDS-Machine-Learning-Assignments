### Exploring Indian restaurant dataset

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy
import seaborn as sns 
```

```python
indian_restaurant = pd.read_csv('restaurants.csv')
indian_restaurant.head()
```

#### Column names

```python
indian_restaurant.columns
```

```python
print(indian_restaurant.isnull().sum())
```

#### Data frame shape and info

```python
indian_restaurant.shape
indian_restaurant.info()
```

```python
indian_restaurant.describe()
```

#### City Count

```python
indian_restaurant['City'].value_counts()
```

```python
indian_restaurant['State'] = indian_restaurant['City']
indian_restaurant['State'] = indian_restaurant['City'].replace({'Bangalore': 'Karnataka', 'Delhi':'Delhi NCR', 'Mumbai':'Maharashtra', 'Kolkata':'Bengal', 'Hyderabad':'Telangana', 'Ahmedabad':'Gujarat', 'Chennai':'Tamil Nadu', 'Pune':'Maharashtra', 'Jaipur':'Rajasthan', 'Chandigarh':'Punjab', 'Indore':'Madhya Pradesh', 'Gurgaon':'Delhi NCR', 'Noida':'Delhi NCR', 'Vadodara':'Gujarat', 'Lucknow':'Uttar Pradesh', 'Agra':'Uttar Pradesh', 'Nagpur':'Maharashtra', 'Surat':'Gujarat', 'Ludhiana':'Punjab', 'Goa':'Goa', 'Ghaziabad':'Delhi NCR', 'Udaipur':'Rajasthan', 'Kochi':'Kerala'})
indian_restaurant['State'].value_counts()
```

### Rating distribution using histogram

```python
plt.figure(figsize=(8,6))
plt.hist(indian_restaurant['Rating'], bins=20, color='green')
plt.title('Distribution of Restaurant Ratings in India')
plt.xlabel('Rating')
plt.ylabel('Count')
plt.show()
```

### Cost distribution

```python
plt.figure(figsize=(8,6))
plt.hist(indian_restaurant['Cost'], bins=20, color='blue')
plt.title('Distribution of Restaurant Costs in India')
plt.xlabel('Cost')
plt.ylabel('Count')
plt.show()
```

### Votes distribution

```python
plt.figure(figsize=(8,6))
plt.hist(indian_restaurant['Votes'], bins=20, color='orange')
plt.title('Distribution of Restaurant Votes in India')
plt.xlabel('Votes')
plt.ylabel('Count')
plt.show()
```

### Distribution of Restaurants across Cities in India

```python
plt.figure(figsize=(15,8))
sns.countplot(y = 'City', data = indian_restaurant, order = indian_restaurant['City'].value_counts().index, palette = 'viridis')
plt.title('Distribution of Restaurants across Cities in India')
plt.xlabel('Number of Restaurants')
plt.ylabel('City')
plt.show()
```
